Minor, Cofactor, Adjoint matrix


```python
A = np.array([[2, 4, 1,], [3, 1, 1], [1, 2, 1]])
print(A)
```

```python
def reduce_matrix(matrix, i,j):
  new_matrix = np.delete(matrix, i, 0)
  new_matrix = np.delete(new_matrix, j, 1)
  return new_matrix
```

```python
cofactor_matrix = np.zeros_like(A, float)

for i in range(A.shape[0]):
  for j in range(A.shape[1]):
    minor = np.linalg.det(reduce_matrix(A, i,j))
    cofactor = ((-1)**(i+j))*minor
    cofactor_matrix[i,j] = cofactor

print(cofactor_matrix)
```

```python
adjoint_matrix = cofactor_matrix.T
inverse_matrix = 1/np.linalg.det(A) * adjoint_matrix
print(inverse_matrix)
print(np.linalg.inv(A))
```
